SELECT AVG(Listaar) AS 'Átlagár',
        MIN(Listaar) AS 'Minimum ár',
        MAX(Listaar) AS 'Maximum ár',
        STDEV(Listaar) AS 'Szórás'
FROM Termek

SELECT CASE WHEN LISTAAR = 0 THEN 'nulla'
                ELSE 'nem nulla' END AS 'nulla-e',
                COUNT(*) AS 'gyakoriság'
FROM Termek
GROUP BY CASE WHEN LISTAAR = 0 THEN 'nulla'
                ELSE 'nem nulla' END
-- A nulla árú termék adathiba lehet

SELECT CASE WHEN LISTAAR - 786.28 \> 1.5_3602.48 THEN 'Kilógó érték' ELSE 'Tipikus érték' END AS 'Kilógó értékek', COUNT(_) aS 'Gyakoriság' FROM Termek GROUP BY CASE WHEN LISTAAR - 786.28 \> 1.5\*3602.48 THEN 'Kilógó érték' ELSE 'Tipikus érték' END -- A kilógó értékeket úgy definiáltam, hogy az átlagtól a szórás másfélszeresére vannak. -- Ami az átlaghoz ennél közelebbre esik, az tipikus értéknek mondható. -- Az árak szórása nagyobb, mint az átlag, és negatív ár nem fordulhat elő, így kilógó értékek csak az átlagnál nagyobbak lehetnek.



SELECT DISTINCT COUNT(MEGNEVEZES) AS 'Termékmegnevezések száma',
                COUNT(TERMEKKOD) AS 'Egyedi termékek száma'
FROM Termek
-- Nincs két azonos nevű termék

SELECT row\_type, COUNT(\*) row\_count FROM (SELECT CASE WHEN LEIRAS IS NULL THEN 'Null' WHEN LEN(LEIRAS) = 0 THEN 'Üres' ELSE 'Nem üres' END AS row\_type FROM Termek) rows GROUP BY row\_type -- A termékek többségéhez nincs leírás



SELECT MIN(FELVITEL) AS 'Legkorábbi',
        MAX(FELVITEL) AS 'Legkésőbbi'
FROM Termek
-- Látható, hogy nincsen gyanúsan régi/új dátum
